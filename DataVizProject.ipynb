```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('inputs/Sleep Log - Ali.csv')
df1
```

```python
# Creating a bubble plot. Demonstrating Daily Productivity rating given hours sleep and sleep quality
x = df1['Total Hours of Sleep'].sort_values(ascending=True)
y = df1['Quality rating'].sort_values(ascending=True)
z = df1['Daily Productivity Rating']


plt.scatter(x, y, s=z*1000, c="red", alpha=0.4)

plt.xlabel("Total Hours of Sleep")
plt.ylabel("Sleep Quality Rating")

plt.show()
```

```python
x = df1['Total Hours of Sleep'].sort_values(ascending=True)
y = df1['Daily Productivity Rating']

plt.scatter(x, y, c="red")

plt.xlabel("Total Hours of Sleep")
plt.ylabel("Daily Productivity Rating")

plt.show()
```

```python
x = df1['Quality rating']
y = df1['Daily Productivity Rating']

plt.scatter(x, y, c="red")

plt.xlabel("Sleep Quality Rating")
plt.ylabel("Daily Productivity Rating")

plt.show()
```

```python
x = df1["Total Hours of Sleep"].sort_values(ascending=True)
y = df1["Daily Productivity Rating"]
std = df1.std(axis=1) # Standard deviation for each data point

# Create a new figure
fig, ax = plt.subplots()

# Plot the data with error bars
ax.errorbar(x, y, yerr=std, fmt='o', capsize=5)

# Add labels and title
ax.set_xlabel("Total Hours of Sleep")
ax.set_ylabel("Daily Productivity Rating")

# Display the plot
plt.show()
```
